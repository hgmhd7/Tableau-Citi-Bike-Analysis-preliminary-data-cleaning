```python
# Import pandas
import pandas as pd
```

```python
# Assign file links to variables
may_17 = "JC-201705-citibike-tripdata.csv"
june_17 = "JC-201706-citibike-tripdata.csv"
july_17 = "JC-201707-citibike-tripdata.csv"

may_18 = "JC-201805-citibike-tripdata.csv"
june_18 = "JC-201806-citibike-tripdata.csv"
july_18 = "JC-201807-citibike-tripdata.csv"

may_19 = "JC-201905-citibike-tripdata.csv"
june_19 = "JC-201906-citibike-tripdata.csv"
july_19 = "JC-201907-citibike-tripdata.csv"
```

```python
# Create DFs
may_17_df = pd.read_csv(may_17)
june_17_df = pd.read_csv(june_17)
july_17_df = pd.read_csv(july_17)

may_18_df = pd.read_csv(may_18)
june_18_df = pd.read_csv(june_18)
july_18_df = pd.read_csv(july_18)

may_19_df = pd.read_csv(may_19)
june_19_df = pd.read_csv(june_19)
july_19_df = pd.read_csv(july_19)
```

```python
# Add a month column to the DFs
may_17_df["month"] = "May"
june_17_df["month"] = "June"
july_17_df["month"] = "July"

may_18_df["month"] = "May"
june_18_df["month"] = "June"
july_18_df["month"] = "July"

may_19_df["month"] = "May"
june_19_df["month"] = "June"
july_19_df["month"] = "July"
```

```python
# Add a month year column to the DFs also just in case 
may_17_df["month year"] = "May 2017"
june_17_df["month year"] = "June 2017"
july_17_df["month year"] = "July 2017"

may_18_df["month year"] = "May 2018"
june_18_df["month year"] = "June 2018"
july_18_df["month year"] = "July 2018"

may_19_df["month year"] = "May 2019"
june_19_df["month year"] = "June 2019"
july_19_df["month year"] = "July 2019"
```

```python
# Stack the DFs to create the summers
summer_17_df = pd.concat([may_17_df, june_17_df, july_17_df])
summer_18_df = pd.concat([may_18_df, june_18_df, july_18_df])
summer_19_df = pd.concat([may_19_df, june_19_df, july_19_df])
```

# Clean and mung the City Bike data for Summer of 2017

```python
# Check DF
summer_17_df.head()
```

```python
# Check DF shape
summer_17_df.shape
```

```python
# Check Dtypes
summer_17_df.dtypes
```

```python
# Exclude nulls in birth year so I can turn the column into integers
summer_17_df_pre_reducd = summer_17_df[~summer_17_df["birth year"].isnull()]

summer_17_df_pre_reducd.head()
```

```python
# Converts the values in the cleaned birth year column to integers
summer_17_df_pre_reducd["birth year"] = summer_17_df_pre_reducd["birth year"].astype(int)

summer_17_df_pre_reducd.head()
```

```python
# Check Dtypes
summer_17_df_pre_reducd.dtypes
```

```python
# Check DF shape
summer_17_df_pre_reducd.shape
```

```python
# Exclude birth years below 1920 and reheck shape
summer_17_df_reduced = summer_17_df_pre_reducd[summer_17_df_pre_reducd["birth year"] >= 1920]

summer_17_df_reduced.shape
```

```python
# Create empty lists to load in genders and ages with the below loop
gender_list_2017 = []
age_list_2017 = []

# Loop through DF to add genders and ages to the above lists
for index, row in summer_17_df_reduced.iterrows():
    
    if row["gender"] == 0:
        gender_list_2017.append("Non-binary/ Unknown")
    elif row["gender"] == 1:
        gender_list_2017.append("Male")
    else:
        gender_list_2017.append("Female")
      
    
    age = 2019-row["birth year"]
    age_list_2017.append(age)
    
    
print(len(gender_list_2017), gender_list_2017[:10], len(age_list_2017), age_list_2017[:10])
```

```python
# Add ages and genders to DF
summer_17_df_reduced["Age"] = age_list_2017
summer_17_df_reduced["Gender"] = gender_list_2017
```

```python
# Add a year column to signify the year and summer
summer_17_df_reduced["Year"] = "Summer 2017"
```

```python
# Add another integer year column just in case
summer_17_df_reduced["Year"] = 2017
```

```python
# Add a trip duration minutes column that converts seconds to minutes
import time

minutes_list_2017 = []

def convert(seconds):
    
    return time.strftime("%H:%M:%S", time.gmtime(seconds)) 



for duration in summer_17_df_reduced["tripduration"]:
    
    new_duration = convert(duration)
    
    minutes_list_2017.append(new_duration)


summer_17_df_reduced["Trip Duration"] = minutes_list_2017
```

```python
# Check DF
summer_17_df_reduced.head(10)
```
